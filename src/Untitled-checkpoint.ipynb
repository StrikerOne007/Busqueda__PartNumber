```python
import re

def corregir_rtx(texto):
    """
    Inserta un espacio entre 'RTX' y un número si están unidos.
    """
    # Expresión regular: Busca 'RTX' seguido de uno o más dígitos
    patron = r'(RTX)(\d+)'
    # Reemplaza con 'RTX ' seguido de los números
    texto_corregido = re.sub(patron, r'\1 \2', texto)
    return texto_corregido

# Ejemplo de uso
ejemplos = [
    "DUAL-RTX4060-O8G",
    "ASUS TUF RTX 4080 16GB",
    "NVIDIA-RTX3080-TI",
    "RTX 4090 TI OC",
    "RTX3090"
]

for texto in ejemplos:
    print(f"Original: {texto} -> Corregido: {corregir_rtx(texto)}")
```

```python

# por valor el id asignado en memoria es diferente, se da en los objetos inmutables

x = 10
def funcion(entrada):
    entrada += 1
    print(id(entrada))

funcion(x)
print(x)
print(id(x))
```

```python
import time

# Test con objeto mutable (lista)
def test_mutable():
    lista = []
    for i in range(100000):
        lista.append(i)
    return lista

# Test con objeto inmutable (string)
def test_inmutable():
    texto = ""
    for i in range(100000):
        texto += str(i)
    return texto

# Medición de tiempo
inicio = time.time()
test_mutable()
tiempo_mutable = time.time() - inicio
print(f"Tiempo con lista (mutable): {tiempo_mutable:.4f} segundos")

inicio = time.time()
test_inmutable()
tiempo_inmutable = time.time() - inicio
print(f"Tiempo con string (inmutable): {tiempo_inmutable:.4f} segundos")
```

```python
import time
import sys

# Creación
lista = [1, 2, 3, 4, 5]
tupla = (1, 2, 3, 4, 5)

# Comparar tamaño en memoria
print(f"Tamaño lista: {sys.getsizeof(lista)} bytes")
print(f"Tamaño tupla: {sys.getsizeof(tupla)} bytes")

# Test de rendimiento en acceso
def test_acceso(estructura, iteraciones):
    inicio = time.time()
    for _ in range(iteraciones):
        _ = estructura[2]
    return time.time() - inicio

# Test de iteración
def test_iteracion(estructura, iteraciones):
    inicio = time.time()
    for _ in range(iteraciones):
        for item in estructura:
            pass
    return time.time() - inicio

# Test de creación
def test_creacion_lista(n):
    inicio = time.time()
    for _ in range(n):
        _ = [1, 2, 3, 4, 5]
    return time.time() - inicio

def test_creacion_tupla(n):
    inicio = time.time()
    for _ in range(n):
        _ = (1, 2, 3, 4, 5)
    return time.time() - inicio

iteraciones = 1000000
print(f"\nTiempo acceso lista: {test_acceso(lista, iteraciones):.4f} segundos")
print(f"Tiempo acceso tupla: {test_acceso(tupla, iteraciones):.4f} segundos")
print(f"\nTiempo iteración lista: {test_iteracion(lista, iteraciones):.4f} segundos")
print(f"Tiempo iteración tupla: {test_iteracion(tupla, iteraciones):.4f} segundos")
print(f"\nTiempo creación lista: {test_creacion_lista(iteraciones):.4f} segundos")
print(f"Tiempo creación tupla: {test_creacion_tupla(iteraciones):.4f} segundos")
```

```python
import turtle

def arbol(longitudRama,t):
    if longitudRama > 5:
        t.forward(longitudRama)
        t.right(20)
        arbol(longitudRama-15,t)
        t.left(40)
        arbol(longitudRama-15,t)
        t.right(20)
        t.backward(longitudRama)

def main():
    t = turtle.Turtle()
    miVentana = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.color("green")
    arbol(75,t)
    miVentana.exitonclick()

main()
```
